# Actividad 3
- Borré los productos que no tenian datos porque no em servirá para hacer un tipo de gráfico

```python
df.dropna(subset=["Producto"], inplace=True)
```

# Capitalizar nombres
- Utilicé str.title() para que los nombres y apellidos sean capitalizados, o sea poner su primer letra en mayúscula.

```python
df["Nombre_Cliente"] = df["Nombre_Cliente"].str.title()
df["Apellido_Cliente"] = df["Apellido_Cliente"].str.title()
```

# Rellenar regiones y métodos de pago
- Decidí mantener los datos nulos como Desconocidos porque igualmente me pueden servir los precios y el producto para un gráfico, utilicé fillna para esto

```python
df["Region"] = df["Region"].fillna("Desconocido")
df["Metodo_Pago"] = df["Metodo_Pago"].fillna("Desconocido")
```

# Formatear fechas
- Decidí crear una serie y 2 df nuevos para manejar los formatos de las fechas que eran invalidas, luego combiné ambos dataframes formateados y se los pasé a la columna fecha del df

```python
mis_fechas = pd.Series(df["Fecha"])

fecha_formateada_1 = pd.to_datetime(mis_fechas, format="%Y-%m-%d", errors="coerce")
fecha_formateada_2 = pd.to_datetime(mis_fechas, format="%d-%m-%Y", errors="coerce")

fecha_formateada = fecha_formateada_1.combine_first(fecha_formateada_2)

df["Fecha"] = fecha_formateada
```

# Rellenar los valores numéricos
- Decidí utilizar fillna para rellenar los NaN en las columnas total venta y precio unitario porque considero que son datos importantes
- Utilicé la media para rellenar los datos
- Usé round para redondear a 2 decimales

```python
df["Precio_Unitario"] = df["Precio_Unitario"].fillna(df["Precio_Unitario"].mean()).round(2)
df["Total_Venta"] = df["Total_Venta"].fillna(df["Total_Venta"].mean()).round(2)
```

## Verificar los valores nulos y tipos de datos

## Actividad 4
- Calculos descriptivos del 
- Primero utilicé describe() para ver las estadísticas descriptivas de mi dataframe
- Agrupé los productos por región para ver la cantidad de productos que hay por región
- Hay 5 productos de las cuales su región es desconocida

```python
# Estas fueron las estadísticas básicas tras solucionar todos lso errores de mi dataframe
print(df.describe())

cantidad_producto_por_region = df.groupby("Region")["Producto"].nunique()
cantidad_producto_por_region
```

## Actividad 5
- Grafiqué los datos agrupados de la cantidad de productos por región en un gráfico de barras

```python
plt.Figure(figsize=(8,4))
plt.bar(cantidad_producto_por_region.index, cantidad_producto_por_region.values, edgecolor="red")
plt.title("Cantidad de productos por región", fontweight="bold")
plt.ylabel("Cantidad de productos")
plt.xlabel("Regiones")
plt.xticks(rotation=30)
plt.grid()
plt.ylim(0,25)
plt.show()
```

```python
import sqlite3

conn = sqlite3.connect("data_ventas.sqlite")

df.to_sql("data_ventas", conn, if_exists="replace", index="False")
```
